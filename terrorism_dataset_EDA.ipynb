### Check Unique Values for each variable.

````python
# Check Unique Values for each variable.

#Calculaing unique values for most continous cols might not give any insight. Sometime I may want to display 
#unique values of only catgorical/object dtypes so I created function here.

def show_unique_of_dtype_cols(df,dtype):
	'''
	description : It gives the unique values of each columns with given dtype.
	parameters : 
    		df : Whole df from which you want to display unique values of cols.
            
    		dtype : 'object' - To get info of categorical cols
    				'number' - To get info of numberical cols
                    'int' - To get info of only integer col
                    'float' - To get info of only float col
	'''
	req_cols = df.select_dtypes(include = dtype)
	for col in req_cols:
		display(df[col].value_counts())
        
        
        
        
#printing unique value of only integer cols
show_unique_of_dtype_cols(df, 'int')

````

````python
#printing only categorical/string dtype cols
show_unique_of_dtype_cols(df, 'object')
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#First and foremost thing is treating missing values.But in this dataset we must say missing columns because most of 
#the columns has almost no values in it

#dropping cols which has missing values more than 5000. I chose 5000 because it can be fillable with rest 1.80+ values.
#Filling cols which has more than 5000 null values,in my opinion may miss approximate values

dataf = df.copy(deep = True)
dataf.drop(df.loc[:,df.isna().sum() > 5000].columns,axis = 1,inplace = True)
print("The total number of columns now " ,dataf.shape[1])
````

````python
#Again I am going to drop some columns here which is not needed for us.
dataf.drop(['target1','natlty1','natlty1_txt','guncertain1','ishostkid','multiple','doubtterr','specificity'],axis = 1, inplace = True)
````

````python
dataf.isna().sum()
````

````python
dataf[dataf['latitude'].isna() & dataf['longitude'].isna()].index
````

````python
#Now I am going to fill the latitude and longitude.
#I am going to fill latitude and longitude if the country and city matches
#this might not be exact match where the incident occured but closest to that. 

for index in dataf[(dataf['latitude'].isna()) & (dataf['longitude'].isna())].index:
    for df_idx in dataf[(dataf['country'] == dataf.loc[index,'country']) & 
             (dataf['city'] == dataf.loc[index,'city'])].index:
        
        if dataf.loc[df_idx,'latitude'] !=np.nan and dataf.loc[df_idx,'longitude'] !=np.nan:
                dataf.loc[index,'latitude'] = dataf.loc[df_idx,'latitude']
                dataf.loc[index,'longitude'] = dataf.loc[df_idx,'longitude']
                break
        else:
            
            continue
````

````python
dataf[['longitude','latitude']].isna().sum()
````

````python
#Now lets fill the city and provstate 
#I am going to use the same method which I used to fill the latitude and longitude

for index in dataf[dataf['provstate'].isna()].index:
    for df_idx in dataf[(dataf['latitude'] == dataf.loc[index,'latitude']) & 
             (dataf['longitude'] == dataf.loc[index,'longitude'])].index:
        
        if dataf.loc[df_idx,'provstate'] !=np.nan:
                dataf.loc[index,'provstate'] = dataf.loc[df_idx,'provstate']
                break
        else:
            
            continue
````

````python
dataf[['provstate']].isna().sum()
````

````python
for index in dataf[dataf['city'].isna()].index:
    for df_idx in dataf[(dataf['latitude'] == dataf.loc[index,'latitude']) & 
             (dataf['longitude'] == dataf.loc[index,'longitude'])].index:
        
        if dataf.loc[df_idx,'city'] !=np.nan:
                dataf.loc[index,'city'] = dataf.loc[df_idx,'city']
                break
        else:
            
            continue
````

````python
dataf[['city']].isna().sum()
````

````python
#thus we tried to fill the missing values as can as possible
#Its time to drop the rows that has missing values


dataf.dropna(inplace = True)
dataf.isna().sum()
dataf.shape
````
